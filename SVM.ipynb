```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
```

```python
from sklearn.datasets import make_circles
x,y=make_circles(100,factor=.1,noise=.1)
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='bwr')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
classifier=SVC(kernel='linear')
classifier.fit(x_train,y_train.ravel())
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
def plot_decision_boundary(classifier, x, y):
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title("SVM Decision Boundary on Training Data")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# Plot on training data
plot_decision_boundary(classifier, x_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_3d(x, y):
    r = np.exp(-(x**2).sum(1))
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    
    sc = ax.scatter3D(x[:, 0], x[:, 1], r, c=y, s=100, cmap='coolwarm', edgecolors='k', alpha=0.8)
    
    ax.set_xlabel('Feature 1', fontsize=14, labelpad=10)
    ax.set_ylabel('Feature 2', fontsize=14, labelpad=10)
    ax.set_zlabel('Radial Transformation', fontsize=14, labelpad=10)
    ax.set_title('3D Scatter Plot with Radial Transformation', fontsize=16)
    
    plt.colorbar(sc, label='Classes')
    ax.grid(True)
    
    plt.tight_layout() 
    plt.show()
```

```python
plot_3d(x,y)
```
